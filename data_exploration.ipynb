# Attributes
## Sort attributes into categories to get an overview

```python
binary_variables = [
    "Sex",
    "AccidentArea",
    "Fault",
    "PoliceReportFiled",
    "WitnessPresent",
    "AgentType",
    "FraudFound_P",
]
```

```python
nominal_variables = [
    "Month",
    "DayOfWeek",
    "Make",
    "DayOfWeekClaimed",
    "MonthClaimed",
    "MaritalStatus",
    "PolicyType",
    "BasePolicy",
    "VehicleCategory",
]
```

```python
interval_variables = [
    "AgeOfVehicle",
    "VehiclePrice",
    "Days_Policy_Accident",
    "Days_Policy_Claim",
    "PastNumberOfClaims",
    "NumberOfSuppliments",
    "AddressChange_Claim",
    "NumberOfCars",
    "AgeOfPolicyHolder",
]
```

```python
ordinal_variables = ["DriverRating"]
```

```python
numerical_columns = [
    col
    for col in car_insur_df.columns
    if col not in nominal_variables
    and col not in binary_variables
    and col not in ordinal_variables
    and col not in interval_variables
]
```

## Explore Correlations

```python
car_insur_df["ClaimDate"] = pd.to_datetime(car_insur_df["ClaimDate"])
car_insur_df["AccidentDate"] = pd.to_datetime(car_insur_df["AccidentDate"])
# Create a subset of the dataframe with only the numerical columns
numerical_data = car_insur_df[numerical_columns + ["FraudFound_P"]]
# Calculate the correlation matrix
correlation_matrix = numerical_data.corr()
# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
# Set the title
plt.title("Correlation Matrix")
# Display the plot
plt.show()
```

## Explore Claim Distribution

```python
months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]
data = car_insur_df["MonthClaimed"].value_counts()  # .loc[months]

# Create subplots based on the "Year" column
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(16, 4), sharey=True)
axs[0].set_ylabel("Number of Claims")
# Plot for each year
for i, year in enumerate(sorted(car_insur_df["Year"].unique())):
    ax = axs[i]
    year_data = car_insur_df[car_insur_df["Year"] == year]
    year_data = year_data["MonthClaimed"].value_counts().loc[months]
    ax.bar(year_data.index, year_data.values)
    ax.set_title(f"Claims per Month - Year {year}")
    ax.set_xlabel("Month")

# Adjust the layout and spacing
plt.tight_layout()

# Display the plot
plt.show()
```

## Explore Combinations and their impact on the Fraud Flag

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the plot style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 8))
sns.displot(
    data=car_insur_df, x="AgeOfPolicyHolder", y="Make", hue="FraudFound_P", aspect=2
)

# Set the title and labels
plt.title("Fraud Found by Age and Car Manufacturer")
plt.xlabel("Age")
plt.ylabel("Car Manufacturer")


# Display the plot
plt.show()
```

## Explore Distributions

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create box plots for claim size and age, comparing fraud and non-fraud cases
plt.figure(figsize=(10, 5))

# Box plot for Claim Size
plt.subplot(1, 2, 1)
sns.boxplot(x="FraudFound_P", y="ClaimSize", data=car_insur_df)
plt.title("Box Plot of Claim Size by Fraud Found")
plt.xlabel("Fraud Found")
plt.ylabel("Claim Size")

# Box plot for Age
plt.subplot(1, 2, 2)
sns.boxplot(x="FraudFound_P", y="DaysAccidentClaim", data=car_insur_df)
plt.title("Box Plot of Age by Fraud Found")
plt.xlabel("Fraud Found")
plt.ylabel("Age")

# Adjust the layout
plt.tight_layout()

# Display the plots
plt.show()
```
